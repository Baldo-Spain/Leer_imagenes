# Lectura en blanco y negro
1)   cv2.imread('image', cv2.IMREAD_GRAYSCALE)
2)   cv2.imread('image', 0)

 Problema: matplotlib no interpreta los colores como esperariamos
 
 Solución: especificar el mapa de color que queremos usar (cmap ='gray') 

```python
# Lectura en blanco y negro

mi_imagen = cv2.imread('White_dot_8.jpg',cv2.IMREAD_GRAYSCALE)

plt.imshow(mi_imagen)
```

```python
# Pero si lo leo como GRAY_SCALE porqué lo veo amarillo y azul.

# Simplemente matplotlib usa diferentes mapas de color para 
# interpretar la intensidad de los pixeles.

# Solucion: Añade ,cmap = 'gray'
plt.imshow(mi_imagen, cmap = 'gray')
```

# Lectura en color
1)   cv2.imread('image', cv2.IMREAD_COLOR)
2)   cv2.imread('image', 1)

Problema: Los colores RGB no se muestran correctamente... estan invertidos

Solución 1: Utilizaremos numpy para invertir el orden de los datos

Solución 2: Utilizaremos el método cv2.COLOR_BGRA2RGBA para convertir

```python
# Lectura en color OPCION 1

# Leo
mi_imagen_color = cv2.imread('emojiRGB.jpg',cv2.IMREAD_COLOR)

# Muestro
plt.imshow(mi_imagen_color)
```

```python
# Lectura en color OPCION 2

#Leo
mi_imagen_color2 = cv2.imread('emojiRGB.jpg',1)

# Muestro
plt.imshow(mi_imagen_color2)
```

```python
# Problema: El color deberia ser RGB (Red, Green, Blue),
# sin embargo se muestra al reves BGR (Blue, Green, Red)

# Solución 1: Utilizar Numpy para invertir las posiciones de los valores de forma que 
# donde es azul pase a ser rojo
# y donde es rojo pase a ser azul.

# Leo y muestro
mi_imagen_color3 = cv2.imread('emojiRGB.jpg',1)
plt.figure(figsize = [20, 10])
plt.subplot(121); plt.imshow(mi_imagen_color3); plt.title('Antes',fontsize = 30)

# Convierto y muestro
mi_imagen_color3 = mi_imagen_color3[:,:,::-1]
plt.subplot(122); plt.imshow(mi_imagen_color3); plt.title('Despues',fontsize = 30)

# Muestro
plt.imshow(mi_imagen_color3)
```

```python
# Solución 2: Utilizo el potencial de cv2 para convertir formatos 
# mediante el método cv2.COLOR_BGRA2RGBA

# Leo y muestro
mi_imagen_color4 = cv2.imread('emojiRGB.jpg',1)
plt.figure(figsize = [20, 10])
plt.subplot(121); plt.imshow(mi_imagen_color4); plt.title('Antes',fontsize = 30)

# Convierto y muestro
mi_imagen_color4 = cv2.cvtColor(mi_imagen_color4, cv2.COLOR_BGRA2RGBA) # Convierto
plt.subplot(122); plt.imshow(mi_imagen_color4); plt.title('Despues',fontsize = 30)

plt.imshow(mi_imagen_color4)
```

# Lectura original
  cv2.imread('image', cv2.IMREAD_UNCHANGED)


OBSERVACION: cv2 lee las imagenes como BGR y por eo cuando trabajamos con color RGB debemos hacer las conversiones pertinentes

```python
# Leo 
mi_imagen_unchanged = cv2.imread('degradado_R2Transparent.png', cv2.IMREAD_UNCHANGED)

plt.figure(figsize = [20, 10])
plt.subplot(121); plt.imshow(mi_imagen_unchanged); plt.title('Antes',fontsize = 30)

# Convierto y muestro
mi_imagen_unchanged = cv2.cvtColor(mi_imagen_unchanged, cv2.COLOR_BGRA2RGBA) # Convierto
plt.subplot(122); plt.imshow(mi_imagen_unchanged); plt.title('Despues',fontsize = 30)

plt.imshow(mi_imagen_unchanged)
```
